```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot([3, 2], [0, 0], 'ro')
plt.plot([-2, -1], [0, 0], 'bo')
```

```python
def perceptron():
  theta = np.array([0])
  theta_0 = 0

  training_data = (np.array([3]), np.array([2]), np.array([-2]), np.array([-1]))
  training_labels = (1, 1, -1, -1)

  for i in range(10):
    for index, point in enumerate(training_data):
      if training_labels[index] * (np.dot(theta.T, training_data[index]) + theta_0) <= 0:
        theta += training_labels[index] * point
        theta_0 += training_labels[index]

        a, b = theta[0], theta_0

        x = np.linspace(-3, 1, 100)
        y = a*x + b
        print("weight and bias", a, b)

      print("point", point)

      plt.plot([3, 2], [0, 0], 'ro')
      plt.plot([-2, -1], [0, 0], 'bo')
      plt.plot(x, y, '-k')

      if b == 0:
        yield(plt.show())
      else:
        yield(None)
```

```python
per = perceptron()

next(per)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot([-3, 0], [1, 2], 'ro')
plt.plot([-2, 0], [-1, 0], 'bo')
```

```python
def perceptron():
  theta = np.array([[0], [0]])
  theta_0 = 0

  training_data = (np.array([[-2], [0]]), np.array([[-1], [0]]), np.array([[-3], [1]]), np.array([[0], [2]]))
  training_labels = (1, 1, -1, -1)

  for i in range(10):
    for index, point in enumerate(training_data):
      if training_labels[index] * (np.dot(theta.T, training_data[index]) + theta_0) <= 0:
        theta += training_labels[index] * point
        theta_0 += training_labels[index]

        a, b, c = theta[0][0], theta[1][0], theta_0

        x = np.linspace(-3, 1, 100)
        y = -a/b*x - c/b
        print("weight and bias", a, b, c)

      print("point", point)

      plt.plot([-3, 0], [1, 2], 'ro')
      plt.plot([-2, 0], [-1, 0], 'bo')
      plt.plot(x, y, '-k')

      if b == 0:
        yield(plt.show())
      else:
        yield(None)
```

```python
per = perceptron()

next(per)
```
